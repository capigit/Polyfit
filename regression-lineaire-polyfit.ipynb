```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/kaggle/input/advertising/Advertising.csv')
```

```python
df.head()
```

```python
df['total_spend'] = df['TV'] + df['radio'] + df['newspaper']
```

```python
df.head()
```

```python
sns.regplot(data=df, x='total_spend', y='sales')
```

```python
X = df['total_spend']
y = df['sales']
```

```python
np.polyfit(X,y,deg=1)
```

```python
potentiel_spend = np.linspace(0,500,100)
```

```python
predict_sales = 0.04868788 * potentiel_spend + 4.24302822
```

```python
sns.scatterplot(x='total_spend', y='sales', data=df)
plt.plot(potentiel_spend,predict_sales, color='red')
```
